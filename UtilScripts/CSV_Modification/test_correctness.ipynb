```python
df['MODEL'] = 'StarCoder'
df['TOTAL_TIME'] = df['TOTAL_TIME'].div(1000)
df.head()
```

```python
def get_scenario(x):
    if "TEST-" in x:
        return x.split('-')[1]
    return x
```

```python
def get_class(x):
    classname = x.split('.')[0]
    if "TEST-" in x:
        classname = classname.split('-')[1]
    return classname
```

```python
def calculate_success(row):
    total_tests = row['TOTAL_TESTS']
    total_success = row['TOTAL_TESTS'] - row['TOTAL_FAILURES'] - row['TOTAL_SKIPPED'] -row ['TOTAL_ABORTED']
    if total_tests == 0:
        return "0.0%"
    return str((total_success/total_tests)*100)+"%"
```

```python
df['SCENARIO'] =  df['SCENARIO'].apply(get_scenario)
df['FILE_NAME'] =  df['FILE_NAME'].apply(get_class)

df['SUCCESS_RATE'] = df.apply(calculate_success, axis=1)
df.head()
```

```python
cols = list(df.columns.values)
cols
```

```python
df = df[['MODEL',
         'SCENARIO',
 'TOKEN',
 'PROJECT_NAME',
 'CLASS_NAME',
 'FILE_NAME',
 'TOTAL_TESTS',
 'TOTAL_ABORTED',
 'TOTAL_FAILURES',
 'TOTAL_SKIPPED',
        'SUCCESS_RATE',
 'TOTAL_TIME']]
```

```python
df.sort_values(by=['SCENARIO', 'TOKEN'], inplace=True)
```
