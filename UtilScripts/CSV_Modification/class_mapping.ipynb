```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/lsiddiqsunny/Documents/Notre_Dame/Research/ICSE23-results/CodeGen/SF110-Results/csv-data/JaCoCo_CodeGen_SF110.csv')
project_df = df[(df['SCENARIO'] == 'original')][['GROUP', 'PACKAGE', 'CLASS']]
project_df.drop_duplicates(inplace=True)
project_df.describe
```

```python
for model in ['Codex','ChatGPT-3.5','CodeGen']:
    for scenario in ["original", "scenario1", "scenario2", "scenario3", "scenario4"]:
        for token in [2000,4000]:
            if model!='Codex' and token==4000:
                continue
            print(model, scenario, token)
            project_df[model+'_'+scenario+'_'+str(token)] = [0 for i in range(len(project_df))]
```

```python
project_df.to_csv('project.csv', index=False)
```

```python
codex_test_df= pd.read_csv('/Users/lsiddiqsunny/Documents/Notre_Dame/Research/ICSE23-results/OpenAI/SF110-Results/csv-data/Test_Correctness_Codex_SF110_updated.csv')
codex_test_df.head()
```

```python
model = 'Codex'
for scenario in ["original", "scenario1", "scenario2", "scenario3", "scenario4"]:
    for token in [2000,4000]:
        codex_df = codex_test_df[(codex_test_df['SCENARIO'] == scenario) & (codex_test_df['TOKEN'] == token)]
        for index, row in codex_df.iterrows():
            file = row['FILE_NAME'].split('_')[1]
            project_df.loc[(project_df['PACKAGE'] == row['CLASS_NAME']) & (project_df['CLASS'] == file), model+'_'+scenario+'_'+str(token)] = 1
```
