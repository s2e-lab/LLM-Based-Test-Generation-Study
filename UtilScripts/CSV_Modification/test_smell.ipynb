```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/lsiddiqsunny/Documents/Notre_Dame/Research/ICSE23-results/StarCoder/SF110-Results/csv-data/Test_Smell_StarCoder_SF110.csv')
df.head()
```

Remove "RelativeTestFilePath" and "RelativeProductionFilePath" columns

```python
df.drop('RelativeTestFilePath', inplace=True, axis=1)
df.drop('RelativeProductionFilePath', inplace=True, axis=1)
df.head()
```

Add Model, Scenario and Token columns

```python
def get_scenario(x):
    return x.split('-')[1]
```

```python
def get_token(x):
    if "2000" in x:
        return "2000"
    else :
        return "4000"
```

```python
df.insert(0, 'Model', 'StarCoder')
df.insert(1, 'scenario', df['App'].apply(get_scenario))
df.insert(2, 'token', df['TestClass'].apply(get_token))
df.head()
```

```python
df.sort_values(by=['scenario', 'token'], inplace=True)
```

```python
df.to_csv('/Users/lsiddiqsunny/Documents/Notre_Dame/Research/ICSE23-results/StarCoder/SF110-Results/csv-data/Test_Smell_StarCoder_SF110_updated.csv', index=False)
```
