```python
path = "/Users/lsiddiqsunny/Documents/Notre_Dame/Research/ICSE23-results/CodeGen/SF110-Results/csv-data/JaCoCo_Results"
```

```python
csv_files = glob.glob('{}/jacoco-*.{}'.format(path,'csv'))
csv_files
```

```python
len(csv_files)
```

```python
import pandas as pd
```

```python
new_df = pd.DataFrame()
for f in csv_files:
    project_name = f.split('/')[-1].split('_')[1]
    scenario = f.split('/')[-1].split('_')[2]
    token = f.split('/')[-1].split('_')[3].split('.')[0]
    print(project_name, scenario, token)
    df = pd.read_csv(f)
    df['GROUP'] = project_name
    df['Scenario'] = scenario
    df['Token'] = token
    new_df = new_df.append(df)
    
    

# df = pd.concat((pd.read_csv(f) for f in csv_files), ignore_index=True)
```

```python
new_df.to_csv("all_codegen_combined.csv", index=False)
```
