```python
import os
import pandas as pd
import numpy as np
```

```python
rootPath = "/Users/lsiddiqsunny/Documents/Notre_Dame/Research/ICSE23-results/StarCoder/HumanEvalJava-Results/target/surefire-reports/"
```

```python
headers = ["MODEL","PROJECT_NAME","CLASS_NAME","FILE_NAME","SCENARIO","TOKEN","TOTAL_TESTS","TOTAL_FAILURES","TOTAL_SKIPPED","TOTAL_ABORTED","TOTAL_TIME"]
```

```python
information = []
```

```python
text_files = [f for f in os.listdir(rootPath) if f.endswith('.txt')]
len(text_files)
```

```python
for text_file in text_files:
    info = []
    model = "StarCoder"
    info.append(model)
    scenario =text_file.split('.')[0]
    info.append(scenario)
    token_size = 2000
    info.append(token_size)
    classname = text_file.split('.')[1]
    info.append(classname)
    print(scenario, classname)
    with open(rootPath+text_file, 'r') as f:
            lines2 = f.readlines()
            targetLine = ""
            for line2 in lines2:
                if "Tests run" in line2:
                    targetLine = line2
                    break

            if len(lines2)>0 and "Tests run" in lines2[-1]:
                res = targetLine.strip().split()
                total_tests = res[2].split(",")[0]
                total_failures = res[4].split(",")[0]
                total_skipped = res[6].split(",")[0]
                total_aborted = res[8].split(",")[0]
                total_time = res[11].split()[0]
                total_success = int(total_tests) - int(total_failures) - int(total_skipped) - int(total_aborted)
                info.append(int(total_tests))
                info.append(int(total_failures))
                info.append(int(total_skipped))
                info.append(int(total_aborted))
                if total_success == 0:
                    info.append("0%")
                else:
                    info.append(str(round(total_success/int(total_tests)*100, 2))+"%")
                info.append(total_time)
                print(total_tests)
                print(total_failures)
                print(total_skipped)
                print(total_aborted)
                print(total_time)
                
            else:
                info.append(0)
                info.append(0)
                info.append(0)
                info.append(0)
                info.append("0%")
                info.append(0)
    if total_tests !=0:
        information.append(info)
        
```

```python
with open(rootPath+"out.txt", "r") as f:
    lines = f.readlines()
    for line in lines:
        info = []
        info.append("StarCoder")
        parts = line.strip().split("/")
        print(parts)

        project_name = parts[1]
        info.append(project_name)
        print(project_name)

        className = ".".join(parts[2].split(".")[:-2]).split("-")[-1]
        info.append(className)
        print(className)

        fileName = parts[2].split(".")[-2]
        info.append(fileName+".java")
        print(fileName)

        if "original" not in fileName and "scenario" not in fileName:
            continue


        scenario = fileName.split("_")[0]
        info.append(scenario)
        print(scenario)

        if "TEST-" in scenario:
            scenario = scenario.split("-")[-1]


        if "2000" in fileName:
            token = "2000"
        elif "4000" in fileName:
            token = "4000"
        info.append(token)
        print(token)

        with open(rootPath+line.strip(), "r") as f2:
            lines2 = f2.readlines()
            targetLine = ""
            for line2 in lines2:
                if "Tests run" in line2:
                    targetLine = line2
                    break

            if len(lines2)>0 and "Tests run" in lines2[-1]:
                res = targetLine.strip().split()
                total_tests = res[2].split(",")[0]
                total_failures = res[4].split(",")[0]
                total_skipped = res[6].split(",")[0]
                total_aborted = res[8].split(",")[0]
                total_time = res[11].split()[0]
                info.append(total_tests)
                info.append(total_failures)
                info.append(total_skipped)
                info.append(total_aborted)
                info.append(total_time)
                print(total_tests)
                print(total_failures)
                print(total_skipped)
                print(total_aborted)
                print(total_time)
                
            else:
                info.append("0")
                info.append("0")
                info.append("0")
                info.append("0")
                info.append("0")
        information.append(info)
```

```python
np_info = np.array(information)
df = pd.DataFrame(np_info)
```
