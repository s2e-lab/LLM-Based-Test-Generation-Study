```python
import os
import shutil
```

```python
source_folder = "/Users/lsiddiqsunny/Downloads/SF110-20130704-src"
```

```python
subfolders = [f.path for f in os.scandir(source_folder) if f.is_dir()]
```

```python
def copy_contents(src_dir, dst_dir):
    # Get the list of files and directories in the source directory
    contents = os.listdir(src_dir)

    # Iterate over the contents and move them to the destination directory
    for item in contents:
        src_path = os.path.join(src_dir, item)
        dst_path = os.path.join(dst_dir, item)
        if os.path.isdir(src_path):
            # If the item is a directory, recursively move its contents to the destination directory
            if os.path.exists(dst_path):
                shutil.rmtree(dst_path)
            shutil.copytree(src_path, dst_path)
            copy_contents(src_path, dst_path)
        else:
            # If the item is a file, move it to the destination directory
            shutil.copy2(src_path, dst_dir)
```

```python
for subfolder in subfolders:
    if "lib" in subfolder:
        continue
    no_project_name = subfolder.split("/")[-1]
    project_name = subfolder.split("/")[-1].split("_")[1]
    if os.path.exists(os.path.join(subfolder, "evosuite-tests")):
        print(project_name, "has evosuite-tests folder")
        copy_contents(os.path.join(subfolder, "evosuite-tests"), os.path.join(source_folder, "src/test/java"))



    if os.path.exists(os.path.join(subfolder, "build.xml")):
        print(project_name, "has build.xml folder")
        shutil.move(os.path.join(subfolder, "build.xml"), os.path.join(subfolder, "build.xml.bak"))
        with open("sample_build.xml") as f:
            data = f.read()
        data = data.replace("NO_PROJECT_NAME", no_project_name)
        data = data.replace("PROJECT_NAME", project_name)
        with open(os.path.join(subfolder, "build.xml"), "w") as f:
            f.write(data)
    
```
